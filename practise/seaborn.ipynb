```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.sepal_width)
```

```python
sns.scatterplot(x = iris.sepal_length, y = iris.petal_length)
```

```python
sns.histplot(iris.sepal_length)
```

```python
sns.displot(iris.sepal_length)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x = tips.total_bill , y = tips.tip)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = tips.total_bill, y = tips.tip, data = tips, hue = 'smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'sex',hue = 'size',palette='dark')
```

```python
tips['day'].value_counts()
```

```python
sns.catplot(data = tips, x = 'day' , y = 'total_bill')
```

```python
sns.catplot(data = tips, x = 'sex', y= 'total_bill')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip')
```

```python
sns.pairplot(tips)
```
